### Основная функция

```python
def tree_func(root, d, width):
    
    pos[root] = (width, d)
    child = tree_dict[root]
    
    if len(child) == 0:
        return 0
    
    elif len(child) == 1:
        return tree_func(child[0], d, width + 1)
    
    else:
        d_right = tree_func(child[0], d, width + 1)
        d_down = tree_func(child[1], d + d_right - 1, width)
        
        return d_right + d_down - 1
```
